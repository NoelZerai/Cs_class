Machine learning is basically Artificial intelligence that uses existing data to develop a predictive model. Machine learning is useful, but it is not useful for every problem. We use it when you can’t write the code, when you can’t code the rules, many tasks can’t be solved using a basic rule-based solution. Many factors may influence the answer. Sometimes the rules depend on too many factors, and they overlap. In such cases where it's hard to write a code that can handle all the possible combinations, we use ML. ML can be used to learn patterns in the data and make predictions based on those patterns. When the problem involves predictions based on those patterns. Pattern recognition means solving problems through recognizing trends and patterns in data. You may use ML when the output variables aren’t so linear. You don't want to use ML when you are trying to solve simple answer problems you can just use simple traditional programs. You don't want to use ML when you are working on small data sets because it will overlift without enough data. You don’t want to use it when resources are limited or there are time restraints. You also don’t want to use ML in an Ina situation when the patterns and relationships in the data are well understood because it wouldn’t give significantly more insight than a simple program in this situation. Also, when the data is already stable you don’t want to use ML because it won’t add much insght.
