```python
#1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
class_survey_df = pd.read_csv('CSCI_1070_survey_results-2.csv')
class_survey_df.head()
```

```python
#2
class_survey_df['country/state'] = np.where(class_survey_df['What state/country were you born in?'] == 'Japan', 'Japan',
                                     np.where(class_survey_df['What state/country were you born in?'] == 'Illnois', 'Illnois',
                                             np.where(class_survey_df['What state/country were you born in?'] == 'Missouri', 'Missouri','USA')))
class_survey_df.head()
```

```python
#3
class_survey_df = class_survey_df.dropna(subset=['How many pets and of which kind do you have?'])
class_survey_df['How many pets and of which kind do you have?'].str.contains('none', '0')
class_survey_df['Dogs'] = np.where(class_survey_df['How many pets and of which kind do you have?'].str.contains('dog'), '1','0')
class_survey_df['Cats'] = np.where(class_survey_df['How many pets and of which kind do you have?'].str.contains('cat'), '1','0')
class_survey_df['Other_pets'] = np.where(class_survey_df['How many pets and of which kind do you have?'].str.contains('tutle'), '1','0')
class_survey_df.head()
```

```python
#4
class_survey_df = class_survey_df.dropna(subset=['What is your favorite type of food?'])
class_survey_df['What is your favorite type of food?']
cuisine = {'pasta' :' Italian',
           'VVietnamese' : 'Vietnam',
           'Mexican' : 'Mexican',
           'pasta' : 'italian',
           'Injera' : 'Eritrean',
           '' : 'None'}
class_survey_df['Favorite Cuisine'] = class_survey_df['What is your favorite type of food?'].map(cuisine)    
class_survey_df.head()
```

```python
#5
Major = {'data science' : 'Computer Science',
         'Nursing' : 'Nursing',
         'comp sci' : 'computer science',
          'International Business (Graduated)' : 'political science',
          'Health Sciences' : 'Health Sciences',
          'Finance' : 'business',
          'Forensic Science' : 'Forensic Science',
          'Biochemistry' : 'Biochemistry',
          }

class_survey_df['Major'] = class_survey_df['What is your major?'].map(Major)
class_survey_df['Major']
```

```python
#6
class_survey_df['How old are you?'] = class_survey_df['How old are you?'].str.replace('18 years old', '18')
class_survey_df['How old are you?'] = class_survey_df['How old are you?'].str.replace('19 turning 20 on Febuary 28', '19')
class_survey_df.head()   
```

```python
#7
plt.hist(class_survey_df['How old are you?'])
plt.xlabel('Age')
plt.ylabel('No. of students')
#most of the students are either 18 or 19 all the students are 18 or older. 
```

```python
#8
survey = class_survey_df.groupby('How many siblings do you have?').count()
survey.reset_index(inplace=True)
fig, ax = plt.subplots()
ax.pie(survey['What is your name?'],labels = survey['How many siblings do you have?'], autopct='%1.0f%%')
#most of the students have 1 sbling and the same amount of students have 3 and 4. 
```

```python
#9
work= class_survey_df.groupby('Are you working while in school?').count()
work.reset_index(inplace=True)
fig, ax = plt.subplots()
ax.pie(work['country/state'],labels = work['Are you working while in school?'], autopct='%1.0f%%')
#most students are working while in school 
```

```python
#10
country = class_survey_df.groupby('What state/country were you born in?').count()
country.reset_index(inplace=True)
fig, ax = plt.subplots()
ax.pie(country['What state/country were you born in?'],labels = survey['country'], autopct='%1.0f%%')
```
